```python
#importing the framework to perform analysis
import pandas as pd
import numpy as np
```

```python
#converting to a data frame
df = pd.read_csv('rawpvr_2018-02-01_28d_1083 TueFri.csv')
```

```python
#displaying the first four rows of the dataset
df.head(4)
```

**TASK 2.I**

```python
#converting to data time format
df['Date'] = pd.to_datetime(df['Date'], infer_datetime_format=True, errors='coerce')
```

```python
#dropping of all null values in the dataset from the date column
df = df.dropna(subset=['Date'])
```

```python
#converting date to day name and hour format
df['Day'] = df['Date'].dt.day_name()
df['Hour'] = df['Date'].dt.hour
```

```python
#filtering the data based on direction, day as specified tuesday and hour 9am

df_filtered = df[(df['Direction Name'] == 'North') &
                 (df['Day'] == 'Tuesday') &
                 (df['Hour'] == 9)]
```

```python
north_lanes = df_filtered['Lane Name'].unique()
```

```python
#defining a function to calculate the descriptive statistics of the data
def calculate_lane_statistics(lane_name, lane_data):
    # Calculating Range
    range_value = lane_data.max() - lane_data.min()

    # Calculating Q1, Q2 (median), Q3
    q1 = np.percentile(lane_data, 25)
    q2 = np.percentile(lane_data, 50)
    q3 = np.percentile(lane_data, 75)

    # Calculating IQR
    iqr = q3 - q1

    # Displaying results
    print(f"Statistics for {lane_name}:")
    print(f"  Range: {range_value}")
    print(f"  Q1 (1st Quartile): {q1}")
    print(f"  Q2 (Median): {q2}")
    print(f"  Q3 (3rd Quartile): {q3}")
    print(f"  IQR (Interquartile Range): {iqr}\n")

# Calculate statistics for each North lane
for lane in north_lanes:
    # Filter the data for each North lane
    lane_data = df_filtered[df_filtered['Lane Name'] == lane]['Speed (mph)']
    calculate_lane_statistics(lane, lane_data)
```

```python
#defining the dataframe for task 3
df_individual = pd.DataFrame(data=df)
```

```python
#displaying the first two rows of the data set
df_individual.head(2)
```

```python
#getting the day name along with hour specifically from the date column
df_individual['Day'] = df_individual['Date'].dt.day_name()
df_individual['Hour'] = df_individual['Date'].dt.hour
```

```python
#filtering according to the day of week that is tuesday here and hour which is 9
df_individual_filtered = df_individual[(df_individual['Day'] == 'Tuesday') & (df_individual['Hour']== 9) ]
```

```python
north_ns = df_individual_filtered[df_individual_filtered['Lane Name'] == 'NB_NS']  # North Near Side (NS)
north_mid = df_individual_filtered[df_individual_filtered['Lane Name'] == 'NB_MID']  # North Middle (MID)
north_os = df_individual_filtered[df_individual_filtered['Lane Name'] == 'NB_OS']  # North Outer Shoulder (OS)
```

```python
south_ns = df_individual_filtered[df_individual_filtered['Lane Name'] == 'SB_NS']  # South Near Side (NS)
south_mid = df_individual_filtered[df_individual_filtered['Lane Name'] == 'SB_MID']  # South Middle (MID)
south_os = df_individual_filtered[df_individual_filtered['Lane Name'] == 'SB_OS']  # South Outer Shoulder (OS)
```

```python
#defining a function to calculate the traffic statistic for each individual lane of the north and then the south
def calculate_traffic_stats(lane_data):
    # Traffic volume is the number of vehicles (i.e., number of rows)
    traffic_volumes = lane_data['Date'].groupby(lane_data['Date'].dt.minute).count()  # Volume by minute

    if traffic_volumes.empty:
        return {'Range': 0, 'Q1': 0, 'Q2 (Median)': 0, 'Q3': 0, 'IQR': 0}

    # Calculate Range, Q1, Q2, Q3, and IQR
    traffic_range = traffic_volumes.max() - traffic_volumes.min()
    q1 = traffic_volumes.quantile(0.25)
    q2 = traffic_volumes.median()  # Q2 is the median
    q3 = traffic_volumes.quantile(0.75)
    iqr = q3 - q1

    return {
        'Range': traffic_range,
        'Q1': q1,
        'Q2 (Median)': q2,
        'Q3': q3,
        'IQR': iqr
    }
```

```python
# Calculate statistics for North lanes
north_ns_stats = calculate_traffic_stats(north_ns)
north_mid_stats = calculate_traffic_stats(north_mid)
north_os_stats = calculate_traffic_stats(north_os)

# Calculate statistics for South lanes
south_ns_stats = calculate_traffic_stats(south_ns)
south_mid_stats = calculate_traffic_stats(south_mid)
south_os_stats = calculate_traffic_stats(south_os)
```

```python
#Displaying the results of each north individual lane and south individual lane
print("North NS Lane Stats: \n", north_ns_stats)
print("North MID Lane Stats: \n", north_mid_stats)
print("North OS Lane Stats: \n", north_os_stats)

print("\nSouth NS Lane Stats: \n", south_ns_stats)
print("South MID Lane Stats: \n", south_mid_stats)
print("South OS Lane Stats: \n", south_os_stats)
```
